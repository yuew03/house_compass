```python
import pickle
objects = []
with (open("./ExampleData/Opt_rmsprop__Dro_0.0__Num_20_20_16__Wei_0_pruned_wts.pkl", "rb")) as openfile:
    while True:
        try:
            objects.append(pickle.load(openfile))
        except EOFError:
            break
```

```python
dense_1 = objects[0][0]
dense_2 = objects[0][2]
dense_3 = objects[0][4]
dense_4 = objects[0][6]
```

```python
print(len(dense_1), dense_1[0].size)
print(len(dense_2), dense_2[0].size)
print(len(dense_3), dense_3[0].size)
print(len(dense_4), dense_4[0].size)
```

```python
nodes = []
```

```python
# for layer_1
layer1_keys = []
for i in range(len(dense_1)):
    label = "i"+str(i)
    layer1_keys.append(label)
    tmp = {}
    tmp["label"] = label
    tmp["layer"] = 1
    nodes.append(tmp)
```

```python
# for layer_2
layer2_keys = []
for i in range(len(dense_2)):
    label = "h"+str(i)
    layer2_keys.append(label)
    tmp = {}
    tmp["label"] = label
    tmp["layer"] = 2
    nodes.append(tmp)
```

```python
# for layer_3
layer3_keys = []
for i in range(len(dense_3)):
    label = "h"+str(i)
    layer3_keys.append(label)
    tmp = {}
    tmp["label"] = label
    tmp["layer"] = 3
    nodes.append(tmp)
```

```python
# for layer_4
layer4_keys = []
for i in range(len(dense_4)):
    label = "h"+str(i)
    layer4_keys.append(label)
    tmp = {}
    tmp["label"] = label
    tmp["layer"] = 4
    nodes.append(tmp)
```

```python
# for output
layer5_keys = []
for i in range(dense_4[0].size):
    label = "o"+str(i)
    layer5_keys.append(label)
    tmp = {}
    tmp["label"] = label
    tmp["layer"] = 5
    nodes.append(tmp)
```

```python
nodes
```

```python
links = []
```

```python
# from layer1 to layer2
for i in range(len(layer1_keys)):
    for j in range(len(layer2_keys)):
        tmp = {}
        tmp["source"] = layer1_keys[i]
        tmp["target"] = layer2_keys[j]
        tmp["weights"] = float(dense_1[i][j])
        links.append(tmp)
```

```python
# from layer2 to layer3
for i in range(len(layer2_keys)):
    for j in range(len(layer3_keys)):
        tmp = {}
        tmp["source"] = layer2_keys[i]
        tmp["target"] = layer3_keys[j]
        tmp["weights"] = float(dense_2[i][j])
        links.append(tmp)
```

```python
# from layer3 to layer4
for i in range(len(layer3_keys)):
    for j in range(len(layer4_keys)):
        tmp = {}
        tmp["source"] = layer3_keys[i]
        tmp["target"] = layer4_keys[j]
        tmp["weights"] = float(dense_3[i][j])
        links.append(tmp)
```

```python
# from layer4 to layer5
for i in range(len(layer4_keys)):
    for j in range(len(layer5_keys)):
        tmp = {}
        tmp["source"] = layer4_keys[i]
        tmp["target"] = layer5_keys[j]
        tmp["weights"] = float(dense_4[i][j])
        links.append(tmp)
```

```python
len(links)
```

```python
jsonfile = {}
jsonfile["nodes"] = nodes
jsonfile["links"] = links
```
